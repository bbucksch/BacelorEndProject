```python
import pandapower as pp
import pandas as pd
import numpy as np
from pandapower.timeseries.output_writer import OutputWriter


# Install pandapower in a new environment by pip install pandapower==2.13.1
net = pp.from_pickle('./aggregated_grid_2018_with_generators_loads_costs_controllers.p')

out_path = "./AC_OPF_Test" #"./AC_OPF"
ow = OutputWriter(net, output_path=out_path, output_file_type=".csv", csv_separator=",")
# ow.remove_log_variable("res_gen", "vm_pu")
```

```python
# SHUT OFF GENERATOR

def shut_off_gen(bus_datapath, out_datapath, t_init, t_total, bus_number):
    bus_data = pd.read_csv(bus_datapath + "/busInputData.csv")
    total_busses = 47

    for t in range(t_init, t_init + t_total):
        row = (t * total_busses) + bus_number
        bus_data.at[row, "Gen MW"] = 0
    
    bus_data.to_csv(out_datapath + "/busInputData.csv", index=False)

    return bus_data
```

```python
new_bus = shut_off_gen("./AC_OPF", "./AC_OPF_Test", t_init=38, t_total=3, bus_number=13)

# new_bus
```
